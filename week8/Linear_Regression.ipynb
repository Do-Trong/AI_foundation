```python
#data
import numpy as np
X = np.array([6.7, 4.6, 3.5, 5.5])
y = np.array([9.1, 5.9, 4.6, 6.7]) #label
X = X.reshape(-1, 1)
y = y.reshape(-1, 1)
print(X)
print(y)
```

```python
from matplotlib import pyplot as plt
plt.scatter(X, y)
```

# O = w*x + b

# loss funtion: Loss = (O - y)2 = (w*x + b - y)2

# -> Loss'(w) = 2(w*x + b)*x
# -> Loss'(b) = 2(w*x + b)

# cp nht h s: x = x - a* f(x)

```python

X = np.hstack((np.ones((4, 1)), X))
print(X)
w = np.array([0.,1.]).reshape(-1,1)
print(w)
```

```python
learning_rate = 0.01
#cp nht h s 100 ln
numOfIteration = 100
cost = np.zeros((numOfIteration,1)) #lu tr d liu sau mi ln cp nht

for i in range(1, numOfIteration):
    r = np.dot(X, w) - y
    cost[i] = 0.5*np.sum(r*r)
    w[0] -= learning_rate*np.sum(r)
    # correct the shape dimension
    w[1] -= learning_rate*np.sum(np.multiply(r, X[:,1].reshape(-1,1)))
    print(cost[i])
```

```python
# biu  lm loss ang gim
plt.scatter(range(3, 100), 100*cost[3:])
```
