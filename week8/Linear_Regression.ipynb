# Gradient descent

1. Khởi tạo giá trị x = x0 tùy ý


2. Gán x = x - a* f’(x) ( a là hằng số không âm ví dụ a = 0.01)


3. Tính lại f(x): Nếu f(x) đủ nhỏ thì dừng lại, ngược lại tiếp tục bước 2


# giá trị dự đoán:   O = w1*x1 + w2*x2 + ...... wn*xn + b
# viết lại: O = W*X + b
trong đó W là ma trân trọng số, X là vector đặc trưng
# loss funtion L(𝐰, b) = (𝑜 − 𝑦)**2



mục đính là đi tìm giá trị nhỏ nhất của hàm loss funtion 

## khi đó ta cập nhận trọng số W bằng công thức trên
# w = w - a* f’(w)
với f’(w) là đạo hàm riêng theo w



# ví dụ

```python
#data
import numpy as np
X = np.array([6.7, 4.6, 3.5, 5.5])
y = np.array([9.1, 5.9, 4.6, 6.7]) #label
X = X.reshape(-1, 1)
y = y.reshape(-1, 1)
print(X)
print(y)
```

```python
from matplotlib import pyplot as plt
plt.scatter(X, y)
```

# O = w*x + b

# loss funtion: Loss = (O - y)2 = (w*x + b - y)2

# -> Loss'(w) = 2(w*x + b)*x
# -> Loss'(b) = 2(w*x + b)

# cập nhật hệ số: x = x - a* f’(x)

```python

X = np.hstack((np.ones((4, 1)), X))
print(X)
w = np.array([0.,1.]).reshape(-1,1)
print(w)
```

```python
learning_rate = 0.01
#cập nhật hệ số 100 lần
numOfIteration = 100
cost = np.zeros((numOfIteration,1)) #lưu trữ dữ liệu sau mỗi lần cập nhật

for i in range(1, numOfIteration):
    r = np.dot(X, w) - y
    cost[i] = 0.5*np.sum(r*r)
    w[0] -= learning_rate*np.sum(r)
    # correct the shape dimension
    w[1] -= learning_rate*np.sum(np.multiply(r, X[:,1].reshape(-1,1)))
    print(cost[i])
```

```python
# biểu đồ làm loss đang giảm
plt.scatter(range(3, 100), 100*cost[3:])
```
